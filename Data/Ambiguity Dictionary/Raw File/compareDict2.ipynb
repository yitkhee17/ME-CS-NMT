```python
import json
with open('/content/drive/MyDrive/PhD/Research/Dictionary/cambridge-english-malaysian.json') as json_file:
    cem_dict = json.load(json_file)
```

```python
cem_dict[0]
```

```python
cem_word_list = []
for c in cem_dict:
  temp = str(c['title']).strip("'[]").split(",")
  for i in range(len(temp)):
    cem_word_list.append(temp[i].strip(" '"))
```

```python
cem_word_list = list(set(cem_word_list))
```

```python
cem_word_list.sort()
```

```python
if 'hairdressing' in cem_word_list:print('6')
```

```python
len(cem_word_list)
```

```python
with open('/content/drive/MyDrive/PhD/Research/Dictionary/ms_MY.dic') as file:
    mn_dict = [line.rstrip().partition('/')[0] for line in file]
```

```python
amb3 = list(set(mn_dict) & set(cem_word_list))
```

```python
len(amb3)
```

```python
amb3.sort()
```

```python
with open('/content/drive/MyDrive/PhD/Research/Dictionary/sorted_amb3_from_dict.txt', 'w') as f:
    for line in amb3:
        f.write("%s\n" % line)
```
